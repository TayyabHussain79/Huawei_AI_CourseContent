**Simple Linear Regression** - Gradient Descent Implementation
* Manual Implementation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.loadtxt('ex2Data/ex2x.dat', dtype=np.float32)
y = np.loadtxt('ex2Data/ex2y.dat', dtype=np.float32)
```

```python
x, y
```

```python
plt.scatter(x, y, marker='x', c='r')
plt.title("Data")
plt.show()
```

```python
x.shape[0]
```

```python
X = np.ones((x.shape[0], 2))
```

```python
X[:, 1] = x
```

```python
X
```

```python
def hypothesis(x, w):
    return np.dot(x, w)
```

```python
def cost(x, y, w):
    return 1/(2*x.shape[0]) * sum(hypothesis(x, w) - y)**2
```

```python
def gradient(x, y, w):
    return 1/(x.shape[0]) * x.T.dot(hypothesis(x, w) - y)
```

```python
w = np.array([0, 0])
costVal = []
epochs = []
for i in range(1, 300):
    costVal.append(cost(X, y, w))
    w = w - 0.0003 * gradient(X, y, w)
    epochs.append(i)
```

```python
w
```

```python
plt.plot(epochs, costVal)
plt.title("Epochs vs. Cost")
plt.xlabel("Epochs")
plt.ylabel("Cost")
plt.show()
```

```python
plt.scatter(x, y, marker='x', c='r', label= ("Data"))
plt.plot(x, hypothesis(X, w), label= ("fitted line"))
plt.title("Data vs. Fitted Line")
plt.legend()
plt.show()
```
